IMPORTING LIBRARIES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

LOADING DATA

```python
df = pd.read_csv(r"C:\Users\Administrator\Desktop\Datasets\Hospital appointment.csv")
df.head(20)
```

DATA CLEANING

```python
df.columns = (
    df.columns.str.replace("[^0-9a-zA-Z_]+", "_", regex=True)
              .str.strip("_")
)
df.columns
```

```python
df["Date_of_Admission"] = pd.to_datetime(df["Date_of_Admission"])
df["Discharge_Date"] = pd.to_datetime(df["Discharge_Date"])

df["length_of_stay"] = (df["Discharge_Date"] - df["Date_of_Admission"]).dt.days
```

```python
df["length_of_stay"]
```

RISK SCORING

```python
def risk_score(row):
    score = 0
    if row['Age'] >= 60:
        score += 2
    if row['Medical_Condition'] in ['Cancer', 'Diabetes', 'Asthma']:
        score += 3
    if row['Test_Results'] == 'Abnormal':
        score += 2
    if row['Admission_Type'] == 'Emergency':
        score += 2
    return score

df['risk_score'] = df.apply(risk_score, axis=1)
```
